```python
import os
import google.generativeai as genai
import gradio as gr
```

```python
model = genai.GenerativeModel('gemini-pro')
```

```python
def diagnose(symptoms):
    prompt = f"""
You are an AI medical assistant. Based on the symptoms: "{symptoms}", do the following:
1. List 2–3 possible conditions.
2. Mention which condition is most likely.
3. Suggest whether to see a doctor or not.
4. Give some basic care advice.

Symptoms: {symptoms}
"""
    response = model.generate_content(prompt)
    return response.text
```

```python
gr.Interface(fn=diagnose,
             inputs=gr.Textbox(label="Enter Symptoms"),
             outputs="text",
             title="AI Medical Diagnosis Assistant").launch()
```
