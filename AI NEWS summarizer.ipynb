```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# Define the prompt template for summarization
template = """
You are a helpful assistant. Summarize the following news article into 3-5 short and clear bullet points:

{text}

Summary:
"""

# Initialize the PromptTemplate with the defined template
prompt = PromptTemplate(input_variables=["text"], template=template)

# Initialize the LLM (Language Model) with the correct parameters
llm = ChatGoogleGenerativeAI(model="models/chat-bison-001", temperature=0.3)

# Create the LLMChain with the LLM and the prompt
summarization_chain = LLMChain(llm=llm, prompt=prompt)

# Example article for summarization
article = """
NASA's Artemis program is set to return astronauts to the Moon by 2025. This mission will mark the first time a female astronaut lands on the Moon.
The program aims to establish a long-term lunar base as a foundation for future Mars exploration.
"""

# Run the chain to summarize the article
summary = summarization_chain.run(article)

# Output the summary
print(summary)
```

```python
llm = GoogleGenerativeAI(model="gemini-pro", google_api_key=os.environ["GOOGLE_API_KEY"])

template = """
Summarize the following news article in 5 bullet points, focusing on the main events, people involved, and impact.

News Article:
{text}
"""

prompt = PromptTemplate(template=template, input_variables=["text"])
chain = LLMChain(llm=llm, prompt=prompt)
```

```python
def summarize_news(news_text):
    return chain.run(text=news_text)

gr.Interface(
    fn=summarize_news,
    inputs=gr.Textbox(lines=15, placeholder="Paste a news article here..."),
    outputs="text",
    title="AI News Summarizer"
).launch()
```
