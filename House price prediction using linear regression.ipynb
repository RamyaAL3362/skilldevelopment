```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# Create sample house data
data = {
    'Size': [1500, 1800, 2400, 3000, 3500, 1300, 1700, 2600, 2200, 1400],
    'Bedrooms': [3, 4, 3, 5, 4, 2, 3, 4, 3, 2],
    'Location': [1, 2, 2, 3, 3, 1, 1, 2, 3, 1],  # Encoded: 1=suburb, 2=city, 3=metro
    'Price': [300000, 400000, 500000, 600000, 650000, 280000, 340000, 520000, 480000, 290000]
}

df = pd.DataFrame(data)
df.head()
```

```python
# Independent variables (features)
X = df[['Size', 'Bedrooms', 'Location']]

# Dependent variable (target)
y = df['Price']
```

```python
# Split data into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Split data into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create model
model = LinearRegression()

# Train model
model.fit(X_train, y_train)
```

```python
# Predict prices
y_pred = model.predict(X_test)

# Evaluate using MSE
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted House Prices")
plt.grid(True)
plt.show()
```
