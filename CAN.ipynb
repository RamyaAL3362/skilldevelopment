```python
import can

# Use a virtual CAN bus interface (for testing/simulation)
bus = can.interface.Bus(bustype='virtual')

# Create a CAN message
msg = can.Message(arbitration_id=0x123,
                  data=[0x11, 0x22, 0x33, 0x44, 0x55, 0x66, 0x77, 0x88],
                  is_extended_id=False)

# Send the message
try:
    bus.send(msg)
    print("Message sent on virtual CAN bus")
except can.CanError:
    print("Message NOT sent")

# Receive the message (blocking mode, waits for message)
received_msg = bus.recv(timeout=1.0)
print("Received message:", received_msg)
```
