```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input, decode_predictions
import numpy as np
import gradio as gr
```

```python
model = MobileNetV2(weights="imagenet")
```

```python
def classify_image(img):
    img = img.resize((224, 224))                  # Resize image to model's input size
    img_array = np.array(img)                     # Convert to numpy array
    img_array = np.expand_dims(img_array, axis=0) # Add batch dimension
    img_array = preprocess_input(img_array)       # Preprocess for MobileNetV2
    preds = model.predict(img_array)              # Get predictions
    decoded = decode_predictions(preds, top=3)[0] # Decode top 3
    return {label: float(score) for (_, label, score) in decoded}
```

```python
gr.Interface(
    fn=classify_image,
    inputs=gr.Image(type="pil"),
    outputs=gr.Label(num_top_classes=3),
    title="Image Classifier (MobileNetV2)"
).launch()
```
