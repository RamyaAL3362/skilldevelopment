```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
```

```python
# Create synthetic classification data
X, y = make_classification(
    n_samples=1000,        # Number of rows (samples)
    n_features=5,          # Total features
    n_informative=3,       # Important features
    n_redundant=0,         # Redundant features
    n_classes=2,           # Two output classes (0 or 1)
    random_state=42        # Seed for consistent results
)
```

```python
# Column names
column_names = ['Age', 'Income', 'SpendingScore', 'DebtRatio', 'Savings']

# Create DataFrame
df = pd.DataFrame(X, columns=column_names)

# Add target column
df['Purchased'] = y

# Show the first 5 rows
df.head()
```

```python
# Save to CSV
df.to_csv("synthetic_customer_data.csv", index=False)
```
